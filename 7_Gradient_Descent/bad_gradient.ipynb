```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# # height (cm)
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T # each row is a point 
# weight (kg)
y = np.array([ 49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68])
# Load our data set
# X = np.array([[0, 1.0, 2.0, 3.0]]).T   #features
# y = np.array([400, 450, 470.0, 500.0])   #target value
# Visualize data 
plt.plot(X, y, 'ko')
plt.axis([140, 190, 40, 90])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def cost(x, y, w, b):
  m = len(x)
  sum = 0
  for i in range(m):
    sum += (w * x[i] + b - y[i]) ** 2
  return (1 / (2 * m) * sum)

def derivative(x, y, w, b):
  m = len(x)
  yh = np.multiply(x, w) + b - y
  dw = np.dot(yh, x) / m
  db = np.sum(yh) / m
  return dw, db

w = 0
b = -30
alpha = 1e-6

integrations = 1000000
for _ in range(integrations):
  dw, db = derivative(X.T[0], y, w, b)
  w -= alpha * dw
  b -= alpha * db
print(w, b)
```

```python
# Trained
x0 = np.linspace(140, 190, 2, endpoint=True)
y0 = w * x0 + b
plt.plot(x0, y0, color = '#27f849', linewidth = 2, label = "Trained model")

# Expected
from sklearn import datasets, linear_model
regr = linear_model.LinearRegression() 
regr.fit(X, y)
w0 = regr.coef_[0]
w1 = regr.intercept_
x0 = np.linspace(140, 190, 2, endpoint=True)
y0 = w0 * x0 + w1
print(w0, w1)

plt.plot(X, y, 'o', color = 'white', mec = 'k', markersize = 8, label = "training data")
plt.plot(x0, y0, color = 'k', linewidth = 2, label = "Expected model")
plt.axis([140, 190, 40, 90])
plt.xlabel('x', fontsize = 14)
plt.ylabel('y', fontsize = 14)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.legend(loc = "best", fontsize = 14)
```
